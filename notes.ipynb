```
let example = [1;2;3;4]

let rec addTwoToEachElement ls = 
    match ls with
        | [] -> []
        | head :: tail -> head + 2 :: addTwoToEachElement tail

addTwoToEachElement example // [3;4;5;6]

// algemene map
let rec map transformed ls = 
    match ls with
        | [] -> []
        | head :: tail -> transformed head :: map transformed tail

map ((*) 2) example // [2;4;6;8]
```

```
let trueExample = [1;2;10;11]
let falseExample = [1;2;-1;4]

let rec isAscending ls = 
    match ls with
        | [] | [_] -> true
        | first :: second :: _ when second <= first  ->  false
        | _ :: tail -> isAscending (tail)

isAscending trueExample // true
isAscending falseExample
```
